```python
# Dependencies
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
#Lists for storing data 
restaurant_name_list = []
price_list = []
rating_list = []
cuisine_list = []
trip_advisor_url_list = []
address_list = []
website_list = []
```

```python
#Trip Advisor URL
url = "https://www.tripadvisor.com/Restaurants-g6787210-Morris_County_New_Jersey.html"
response = requests.get(url)
nextpage_link = True

#Loop for all of the restaurants' names
while (nextpage_link):
    
    soup = BeautifulSoup(response.text,"html.parser")
    
    # retrieve list of names of restaurant on the page
    reaturant_names = soup.find_all('div',class_= "wQjYiB7z")
   
    # for each name in the list
    for result in reaturant_names:
        try:
            # find the name and the trip advisor url for the restaurant
            restaurant = result.span.a.text
            link = "https://www.tripadvisor.com" + result.span.a['href']
            print(restaurant)
        except AttributeError as e:
            print(e)
            restaurant = ""
            link = ""
        # append the name and url to lists
        restaurant_name_list.append(restaurant)
        trip_advisor_url_list.append(link)
        
    # Cuisine Type         
    cuisine = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine:
        try:
            # find the cuisine type
            cuisine_type = result.span.span.text
        except AttributeError as e:
            cuisine_type = ""
        # append the cuisine to list
        cuisine_list.append(cuisine_type)
                
    # Price Level              
    price = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in price:
        try:
            # find the price
            price_class = result.find_all('span', class_ = 'EHA742uW')
            if len(price_class)>1:
                price_level = price_class[1].find('span', class_ = '_1p0FLy4t').text
            else:
                price_level = ''
        except AttributeError as e:
            print(e)
            price_level = ''
        # append the price to list
        price_list.append(price_level)

    # go to the next page
    next_page = soup.find('a', class_='nav next rndBtn ui_button primary taLnk')
    if next_page:        
        url = "https://www.tripadvisor.com/" + next_page['href']
        print(url)
        response = requests.get(url)
        #time.sleep(1)
    else:
        nextpage_link = False
```

```python
# Search through all the trip advisor restaurant urls to find address and rating
for url in trip_advisor_url_list:
    print(url)
    response = requests.get(url)
    soup_2 = BeautifulSoup(response.text,"html.parser")
    
    #Extract Address
    raw_address = soup_2.find_all('span', class_ ='_2saB_OSe')
    print(raw_address)
    try:
        # find the address
        if not raw_address:
            address = ""
        else:
            address = raw_address[0].text
        print(address)
    except AttributeError as e:
        print(e)
        address = ""
    
    #Extract Review
    rating_class = soup_2.find('svg', class_ = "_3KcXyP0F")['title']
    try:
        # find the rating
        rating_str = rating_class.split(' ')
        if len(rating_str) > 0:
            rating = rating_str[0]
        else:
            rating = ""
        print(rating)
    except AttributeError as e:
        print(e)
        rating = ''
        
    # append the rating and address to list
    rating_list.append(rating)
    address_list.append(address)
```

```python
#Table with everything
overall = {'restaurant_name' : restaurant_name_list,
           'price_level' : price_list,
           'restaurant_rating' : rating_list,
           'cuisine' : cuisine_list,
           'trip_advisor_url' :trip_advisor_url_list,
           'address' :address_list,
           }
df = pd.DataFrame(overall)
df
```
