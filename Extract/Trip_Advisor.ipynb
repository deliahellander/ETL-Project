```python
# Dependencies
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
# Setup splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Trip Advisor URL

url = "https://www.tripadvisor.com/Restaurants-g6787210-Morris_County_New_Jersey.html"

browser.visit(url)
```

```python
#Lists for storing data 
name_list = []
price_list = []
rating_list = []
cuisine_list = []
trip_advisor_list = []
address_list = []
website_list = []


           
           
           
           
          
```

```python
html = browser.html
soup = BeautifulSoup(html,"html.parser")
    
names = soup.find_all('div',class_= "wQjYiB7z")
print(names)
```

```python
#Loop for all of the restaurants' names
for x in range(39):
    
    soup = BeautifulSoup(html,"html.parser")
    
    names = soup.find_all('div',class_= "wQjYiB7z")
   
    for result in names:
        try:
        
        
            restaurant = result.span.a.text
            link = "https://www.tripadvisor.com" + result.span.a['href']
        
            print(restaurant)
            name_list.append(restaurant)
            trip_advisor_list.append(link)
    
        except AttributeError as e:
            print(e)
#Cuisine Type         
    cuisine = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
   
    for result in cuisine:
        try:
            cuisine_type = result.span.span.text

            print(cuisine_type)
            cuisine_list.append(cuisine_type)   

        except AttributeError as e:
                print(e)
                

#Cuisine Price                
    cuisine_price = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    
   
    for result in cuisine_price:  
        try:
            t = result.find_all('span', class_ = 'EHA742uW')
            if len(t)>1:
                print(len(t))
                price_level = t[1].find('span', class_ = '_1p0FLy4t').text
                
            else:
                price_level = ''
            print(price_level)
            price_list.append(price_level)

     


        except AttributeError as e:
            print(e)
        
            
    try:
        
        browser.click_link_by_partial_text('Next')
        time.sleep(1)
        html = browser.html
        
        
    except:
        print("Done")
    
    
    
   
   
```

```python
for url in trip_advisor_list:
    response = requests.get(url)
    soup_2 = BeautifulSoup(response.text,"html.parser")

#Extract Address
    raw_address = soup_2.find_all('span', class_ ='_2saB_OSe')
    raw_address[0].text


    try:

        address = raw_address[0].text


        print(address)

    except AttributeError as e:
         print(e)
            
    new_review = soup_2.find('a', class_ = "_2mn01bsa")


    try:
#Extract Review      
        review_number = new_review.text
        
      
        print(new_review)
    
    except AttributeError as e:
         print(e)
#Extract Website        
    rest_span = soup_2.find_all('a',class_='_2wKz--mA _27M8V6YV')
    rest_url = rest_span[1]['href']
    print(rest_url) 
  
   
```

```python
#Table with everything
overall = {'restaurant_name' : name_list,
           'price_level' : price_list,
           'restaurant_rating' : rating_list,
           'cuisine' : cuisine_list,
           'trip_advisor_url' :trip_advisor_list,
           'address' :address_list,
           'restaurant_website' : website_list}
df = pd.DataFrame(overall)
df
```
